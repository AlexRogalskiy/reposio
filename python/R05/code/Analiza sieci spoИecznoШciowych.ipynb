```python
import community # Modu≈Ç community s≈Çu≈ºy do wykrywania spo≈Çeczno≈õci i poziomu modularno≈õci

G = nx.powerlaw_cluster_graph(100, 1, .4, seed=101)
partition = community.best_partition(G)

for i in set(partition.values()):
   print ("Spo≈Çeczno≈õƒá ", i)
   members = list_nodes = [nodes for nodes in partition.keys() if partition[nodes] == i]
   print (members)

values = [partition.get(node) for node in G.nodes()]
nx.draw_spring(G, cmap = plt.get_cmap('jet'), node_color = values, node_size=30, with_labels=False)
plt.show()

print ("Poziom modularno≈õci:", community.modularity(partition, G))
```

# Wczytywanie graf√≥w, zapisywanie ich w plikach i pobieranie z nich podpr√≥bek

```python
dump_file_base = "dumped_graph"

# Usuwa okre≈õlony plik
def remove_file(filename):
    import os
    if os.path.exists(filename):
        os.remove(filename)
```

```python
G = nx.krackhardt_kite_graph()
```

```python
# Zapis i odczyt plik√≥w w formacie GML
GML_file = dump_file_base + '.gml'
remove_file(GML_file)

nx.write_gml(G, GML_file)
G2 = nx.read_gml(GML_file)

assert(G.edges() == G2.edges())
```
